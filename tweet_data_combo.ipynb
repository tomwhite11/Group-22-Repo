```python
rows=[]
CountDict=[]
for JSON_FILE in JSON_FILES:
    with open(JSON_FILE, "r") as f:
        count=0
        for line in f:
            data = json.loads(line)

            screen_name = data["user"]["screen_name"]
            tweet_id = data["id_str"]
            reply_count = data["reply_count"]
            quote_count = data["quote_count"]
            retweet_count = data["retweet_count"]
            like_count = data["favorite_count"]
            if "sentiment" in data["entities"]:
                Negative_tweet_sentiment=data["entities"]["sentiment"]["probabilities"]["Very negative"]+data["entities"]["sentiment"]["probabilities"]["Negative"]
                Positive_tweet_sentiment=data["entities"]["sentiment"]["probabilities"]["Very positive"]+data["entities"]["sentiment"]["probabilities"]["Positive"]
            else:
                Negative_tweet_sentiment = None;
                Positive_tweet_sentiment = None;

            if "extended_tweet" in data:
                text = data["extended_tweet"]["full_text"]
            else:
                text = data["text"]

            if "deep state" and "interference" and "antivax" in text:
                rows.append((tweet_id, screen_name, text, Negative_tweet_sentiment, Positive_tweet_sentiment, reply_count, quote_count, retweet_count, like_count))
                count=count+1
        CountDict.append([JSON_FILE[62:72],count])
```

```python
df = pd.DataFrame(data=rows, columns=["tweet_id", "screen_name", "tweet_text","negative_tweet_sentiment","positive_tweet_sentiment","reply_count","quote_count","retweet_count","like_count"])
CountDf = pd.DataFrame(data=CountDict, columns=["date", "count"])
print(CountDf)
```
